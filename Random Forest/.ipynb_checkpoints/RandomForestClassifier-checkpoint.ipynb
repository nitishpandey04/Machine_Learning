```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Importing the dataset

```python
df = pd.read_csv("heart_v2.csv")
```

```python
df.head()
```

#### Splitting the train and test dataset

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, train_size=0.7)
```

```python
df.shape, train_df.shape, test_df.shape
```

```python
x_train = train_df.drop(columns="heart disease")
y_train = train_df["heart disease"]

x_test = test_df.drop(columns="heart disease")
y_test = test_df["heart disease"]
```

#### Making convinience functions to plot and evaluate the decision trees

```python
from IPython.display import Image
from six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
```

```python
def plot(dt_classifier):
    dot_data = StringIO()

    export_graphviz(dt_classifier, out_file=dot_data, filled=True, rounded=True, feature_names=x_train.columns,
                   class_names=["No disease", "Disease"])

    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

    return Image(graph.create_png())
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

def evaluate(dt_classifier):
    y_train_pred = dt_classifier.predict(x_train)
    y_test_pred = dt_classifier.predict(x_test)
    print("Train set performance")
    print("Acc-", accuracy_score(y_train, y_train_pred))
    print(confusion_matrix(y_train, y_train_pred))
    print("-"*25)
    print("Test set performance")
    print("Acc-", accuracy_score(y_test, y_test_pred))
    print(confusion_matrix(y_test, y_test_pred))
```

#### Builing the random forest classifier on small hyperparameters

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=3, n_jobs=-1, n_estimators=10, oob_score=True)
```

```python
rf.fit(x_train, y_train)
```

```python
rf.feature_importances_
```

```python
rf.bootstrap
```

```python
rf.estimators_
```

```python
rf.get_params()
```

```python
rf.oob_score_
```

```python
tree0 = rf.estimators_[0]
plot(tree0)
```

#### The first tree of the forest is plotted above

```python
plot(rf.estimators_[1])
```

#### The second tree is so different than the first tree

```python
evaluate(rf)
```

#### Random Forest shows much better generalization than individual Decision Trees and we have not even done hyperparameter tuning!

#### Hyperparameter Tuning

```python
rf.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    "max_depth":[3, 4, 5, 6, 7, 8],
    "max_features":[2, 3],
    "min_samples_leaf":[5, 10, 15, 20, 25],
    "min_samples_split":[10, 15, 20, 25, 30],
    "n_estimators":[20, 30, 40, 50, 60]
}

rf_basic = RandomForestClassifier(n_jobs=-1)

gs = GridSearchCV(estimator=rf_basic,
                 param_grid=params,
                 scoring="accuracy",
                 n_jobs=-1,
                 cv=4,
                 verbose=1)
```

```python
gs.fit(x_train, y_train)
```

```python
rf_best = gs.best_estimator_
```

```python
gs.best_params_
```

```python
evaluate(rf_best)
```

#### Let's set the n_estimators to fixed value of 100 because in random forest, more the trees better the forest

```python
params = {
    "max_depth":[3, 5, 10, 15, 20],
    "max_features":[2, 3],
    "min_samples_leaf":[5, 10, 20, 30, 40, 50],
    "min_samples_split":[10, 20, 40, 60, 80]
}

rf_basic = RandomForestClassifier(n_jobs=-1, n_estimators=100)

gs = GridSearchCV(estimator=rf_basic,
                 param_grid=params,
                 scoring="accuracy",
                 n_jobs=-1,
                 cv=4,
                 verbose=2)
```

```python
gs.fit(x_train, y_train)
```

```python
rf_best = gs.best_estimator_
```

```python
evaluate(rf_best)
```
