```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
pd.options.display.max_columns = 100
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, train_size=0.8, random_state=0, stratify=df["Attrition_Yes"])
```

```python
train_df.shape, test_df.shape
```

```python
x_train = train_df.drop(columns="Attrition_Yes")
y_train = train_df.Attrition_Yes

x_test = test_df.drop(columns="Attrition_Yes")
y_test = test_df.Attrition_Yes
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adb = AdaBoostClassifier(n_estimators=100)
```

```python
adb.fit(x_train, y_train)
```

```python
y_train_pred = adb.predict(x_train)
```

```python
y_test_pred = adb.predict(x_test)
```

```python
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, plot_roc_curve)
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
print(classification_report(y_train, y_train_pred))
```

```python
plot_roc_curve(adb, x_train, y_train)
plt.show()
```

```python
plot_roc_curve(adb, x_test, y_test)
plt.show()
```

It is showing some overfitting, let's reduce the learning rate to 0.5

```python
adb = AdaBoostClassifier(n_estimators=100, learning_rate=0.5)

adb.fit(x_train, y_train)

y_train_pred = adb.predict(x_train)

y_test_pred = adb.predict(x_test)
```

```python
plot_roc_curve(adb, x_train, y_train)
plt.show()
```

```python
plot_roc_curve(adb, x_test, y_test)
plt.show()
```

#### Let's do some hyperparameter tuning using GridSearchCV to find the optimal values

```python
from sklearn.model_selection import GridSearchCV

params = {
    "n_estimators":[50, 100, 150, 200, 250],
    "learning_rate":[0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
}

adb_basic = AdaBoostClassifier()

gs = GridSearchCV(estimator=adb_basic,
                 param_grid=params,
                 scoring="accuracy",
                 n_jobs=-1,
                 cv=4)
```

```python
gs.fit(x_train, y_train)
```

```python
adb_best = gs.best_estimator_
```

```python
plot_roc_curve(adb_best, x_train, y_train)
plt.show()
```

```python
plot_roc_curve(adb_best, x_test, y_test)
plt.show()
```
