```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

### Data Preparation

```python
df.Attrition = df.Attrition.apply(lambda x : 1 if x == "Yes" else 0)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, train_size=0.75)
```

```python
train_df.shape, test_df.shape
```

#### Train Test Split

```python
x_train = train_df.drop(columns="Attrition")
y_train = train_df.Attrition

x_test = test_df.drop(columns="Attrition")
y_test = test_df.Attrition
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbm = GradientBoostingClassifier()

gbm.fit(x_train, y_train)
```

```python
gbm.get_params()
```

```python
from IPython.display import Image
from six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
import graphviz

def plot(dt_classifier):
    dot_data = StringIO()

    export_graphviz(dt_classifier, out_file=dot_data, filled=True, rounded=True, feature_names=x_train.columns,
                   class_names=["No Attrition", "Attrition"])

    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

    return Image(graph.create_png())
```

```python
plot(gbm.estimators_[0][0])
```

#### n_estimators=100 will train 100 decision trees. The first one is plotted above. 

- each tree trains to predict the residuals of the previous tree
- the prediction of this tree multiplied by learning rate is the contribution of this tree to the overall prediction

#### In both random forest and gradient boosting, we use lots of fully grown trees, the difference is in the ensemble technique which is bagging and boosting respectively.

#### We won't be doing hyperparameter tuning here.
