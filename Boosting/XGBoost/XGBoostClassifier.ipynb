```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
pd.options.display.max_columns=100
```

```python
df.head()
```

### Data Preparation

```python
df.Attrition = df.Attrition.apply(lambda x : 1 if x == "Yes" else 0)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, train_size=0.75)
```

```python
x_train = train_df.drop(columns="Attrition")
y_train = train_df.Attrition

x_test = test_df.drop(columns="Attrition")
y_test = test_df.Attrition
```

```python
x_train.head()
```

### Modelling

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, plot_roc_curve, accuracy_score
```

```python
accuracy_score(y_train, model.predict(x_train))
```

```python
accuracy_score(y_test, model.predict(x_test))
```

#### The model is overfitting because it is showing 100% accuracy on train data and only 86% on test data.

### Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

#### RandomSearchCV

```python
params = {
    "learning_rate" : [0, 0.1, 0.2, 0.3],
    "reg_alpha" : [0, 0.1, 0.2, 0.3],
    "reg_lambda" : [0, 0.1, 0.2, 0.3]
}

model_basic = xgb.XGBClassifier()

rscv = RandomizedSearchCV(estimator=model_basic,
                         param_distributions=params,
                         scoring="accuracy",
                         cv=4,
                         n_jobs=-1)
```

```python
rscv.fit(x_train, y_train)
```

```python
model_best = rscv.best_estimator_
```

```python
model_best.get_params()
```

```python
accuracy_score(y_train, model_best.predict(x_train))
```

```python
accuracy_score(y_test, model_best.predict(x_test))
```

#### There is a slight improvement in the model performance due to RandomizedSearchCV.

#### GridSearchCV

```python
params = {
    "learning_rate" : [0, 0.1, 0.2, 0.3],
    "reg_alpha" : [0, 0.1, 0.2, 0.3],
    "reg_lambda" : [0, 0.1, 0.2, 0.3]
}

model_basic = xgb.XGBClassifier()

gscv = GridSearchCV(estimator=model_basic,
                   param_grid=params,
                   scoring="accuracy",
                   n_jobs=-1,
                   cv=4)
```

```python
gscv.fit(x_train, y_train)
```

```python
model_best = gscv.best_estimator_
```

```python
accuracy_score(y_train, model_best.predict(x_train))
```

```python
accuracy_score(y_test, model_best.predict(x_test))
```

#### GridSearchCV also found the same set of hyperparameters the best one.
